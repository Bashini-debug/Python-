```python
# Step 2 : load the excel data

import pandas as pd

df = pd.read_excel('DA -Task 2..xlsx')  # Use the exact filename


```

```python
# Step 3: Preview the data

df.head()
```

```python
#Step 4: Select the top 5 critical columns that might be most insightful for stakeholders according to your data understanding

critical_cols = [
    'GLOBAL_LABOR_CODE_DESCRIPTION',
    'REPORTING_COST',
    'CORRECTION_VERBATIM',
    'BODY_STYLE',
    'REPAIR_AGE'
]
df_alt = df[critical_cols]
```

```python
#Step 5: To Generate Visualization

# 1) Box Plot: Repair Cost Distribution by Vehicle Age

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.barplot(data=df, x='REPAIR_AGE', y='REPORTING_COST')
plt.title('Repair Cost Distribution by Vehicle Age')
plt.show()

```

```python
# 2) Bar Plot: Average Reporting Cost by Repair Type

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.barplot(data=df, x='GLOBAL_LABOR_CODE_DESCRIPTION', y='REPORTING_COST', estimator='mean')
plt.xticks(rotation=90)
plt.title('Average Reporting Cost by Repair Type')
plt.tight_layout()
plt.show()
```

```python
# 3) Count Plot: Frequency of Body Styles
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.countplot(data=df, x='BODY_STYLE', order=df['BODY_STYLE'].value_counts().index)
plt.title('Repair Frequency by Body Style')
plt.xticks(rotation=45)
plt.show()
```
